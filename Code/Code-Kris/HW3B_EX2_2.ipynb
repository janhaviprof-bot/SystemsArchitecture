```python
import pandas as pd
import numpy as np

# 1. Load the fully constrained architectural space from Exercise 1
# This represents the "whittled down" full factorial matrix.
df = pd.read_csv('arch_space.csv')

# Verify the global space size is exactly 184,212
N_total = len(df)
print(f"Total valid architectures loaded: {N_total}")

target_sample_size = 90 # Target size < 100

# 2. Calculate the deterministic stride (k)
k = N_total // target_sample_size

# 3. Sort lexicographically to ensure uniform categorical grouping
# This orders the dataframe logically from D1 -> D8
columns_to_sort = list(df.columns)
df_sorted = df.sort_values(by=columns_to_sort).reset_index(drop=True)

# 4. Perform Systematic Sampling (Deterministic Grid Extraction)
deterministic_sample = df_sorted.iloc[::k].head(target_sample_size)

# 5. Format into vector representation strings
# Vector structure: D1-D2-D3-D4-D5-D6-D7-D8
vectors = []
for _, row in deterministic_sample.iterrows():
    # Concatenate D4 bits
    d4_str = f"{int(row['d4_1'])}{int(row['d4_2'])}{int(row['d4_3'])}{int(row['d4_4'])}{int(row['d4_5'])}{int(row['d4_6'])}"
    # Concatenate D7 digits
    d7_str = f"{int(row['d7_1'])}{int(row['d7_2'])}{int(row['d7_3'])}{int(row['d7_4'])}{int(row['d7_5'])}{int(row['d7_6'])}"

    # Build final vector
    vec = f"{int(row['d1'])} - {int(row['d2'])} - {int(row['d3'])} - {d4_str} - {int(row['d5'])} - {int(row['d6'])} - {d7_str} - {int(row['d8'])}"
    vectors.append(vec)

print(f"\n--- Generated {len(vectors)} Deterministic Architectures ---")
for i, v in enumerate(vectors):
    print(f"Arch {i+1:02d}: {v}")

# Save to CSV for submission
pd.DataFrame(vectors, columns=['Vector_Representation']).to_csv('deterministic_sample.csv', index=False)
```
