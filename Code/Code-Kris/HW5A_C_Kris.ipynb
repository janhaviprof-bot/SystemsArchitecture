```python
import numpy as np

def compute_entropy(probabilities):
    """
    Computes the Shannon entropy of a probability mass function.
    Assumes base 2 for units of 'bits'.
    """
    p = np.array(probabilities)
    # Filter out zero probabilities to avoid log2(0) errors
    p = p[p > 0]
    return -np.sum(p * np.log2(p))

# Define the Probability Mass Functions (PMFs)
p_X = np.array([0.1, 0.2, 0.4, 0.2, 0.1])
q_Y = np.array([1/5, 1/15, 1/3, 2/15, 4/15])

# Compute individual entropies
H_X = compute_entropy(p_X)
H_Y = compute_entropy(q_Y)

# Compute Mutual Information I(X;Y)
# By definition, if X and Y are independent, their joint probability
# P(X,Y) = P(X)P(Y). Therefore, the mutual information is strictly 0.
I_XY = 0.0

# Print results
print(f"Entropy of X, H(X) = {H_X:.4f} bits")
print(f"Entropy of Y, H(Y) = {H_Y:.4f} bits")
print(f"Mutual Information, I(X;Y) = {I_XY:.4f} bits")
```
